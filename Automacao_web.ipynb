- Importar planilha pro Python
- Percorrer planilha
    - para cada linha:
        - pegar preco Amazon
        - pegar preco lojas americanas
        - pegar preco magalu
    - registrar o menor preco
    - registrar local do menor preco
- Verifica se o desconto é de mais de 20% e envia email
- Salvar a planilha

```python
import pandas as pd
from selenium import webdriver

produtos = pd.read_excel(r'Produtos.xlsx')
#para substituir o NaN por um traço
produtos = produtos.fillna('-')

display(produtos)


```

```python
def transforma_texto(texto):
    return float(texto.replace('R$','').replace('.','').replace(',','.'))
```

```python
import time

driver = webdriver.Chrome()
lista_precos = {}
dict_vazio = {}
enviar_email = False
desconto = 0.1

for i, linha in produtos.iterrows():

    # Entrar no site da amazon, descobrir o objeto onde o preço está e armazenar numa variavel
    if (linha['Amazon'] != 'False'):
        driver.get(linha['Amazon'])
        try: 
            preco_amazon = driver.find_element_by_xpath('//*[@id="price_inside_buybox"]').text
        except:
            preco_amazon = driver.find_elemento_by_css_selector('#olp_feature_div span.a-size-base')
        preco_amazon = transforma_texto(preco_amazon)
        
        lista_precos.update({preco_amazon:'Amazon'})
        
        print(f'Amazon - {linha["Link Produto"]}: {preco_amazon}')

    # pegar preco lame
    driver.get(linha['Lojas Americanas'])
    try:
        preco_lame = driver.find_element_by_class_name('src__BestPrice-sc-1jvw02c-5').text
    except:
        time.sleep(10)
        preco_lame = driver.find_element_by_class_name('src__BestPrice-sc-1jvw02c-5').text
    preco_lame = transforma_texto(preco_lame)
    lista_precos.update({preco_lame:'Lojas Americanas'})
    print(f'Lojas Americanas - {linha["Link Produto"]}: {preco_lame}')

    #pegar preco mglu
    driver.get(linha['Magazine Luiza'])
    preco_mglu = driver.find_element_by_class_name('price-template__text').text
    preco_mglu = transforma_texto(preco_mglu)
    lista_precos.update({preco_mglu:'Magazine Luiza'})
    print(f'Magazine Luiza - {linha["Link Produto"]}: {preco_mglu}')
    
    lista_precos = sorted(lista_precos.items())
    
    produtos.loc[i, 'Preço Atual'] = lista_precos[0][0]
    produtos.loc[i, 'Local'] = lista_precos[0][1]
    
    preco_original = linha['Preço Original']
    
    if lista_precos[0][0] <= preco_original*(1-desconto)
        enviar_email = True
        
    lista_precos = dict_vazio
    

print(lista_precos)    

driver.quit()
```
